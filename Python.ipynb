```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
covid_data = pd.read_csv("C:/Covid-19.csv")
covid_data
```

```python
Thailand = covid_data['geoId']=='TH'
dataTH = covid_data[Thailand]
case_TH = dataTH['cases'] # cases in Thailand
total_TH = np.sum(case_TH)
print('Total case in thailand is: ',total_TH)
```

```python
death_TH = dataTH['deaths'] # deaths in Thailand
totald_TH = np.sum(death_TH)
print('Total death case in thailand is: ',totald_TH)
```

```python
TH_pop = 69428528
caseper = (total_TH/TH_pop)*100
print('Percent of Thailand case: ',caseper)
```

```python
deathper = (totald_TH/total_TH)*100
print('Percent of Thailand death case: ',deathper)
```
