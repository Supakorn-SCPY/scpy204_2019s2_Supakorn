```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astroquery.sdss import SDSS
```

```python
query = 'SELECT \
   s.z as redshift, s.zErr,p.petroMag_g, p.petroMagErr_g \
FROM PhotoObjALL AS p \
   JOIN SpecObjALL AS s ON s.bestobjid = p.objid \
WHERE \
   zWarning = 0 \
   AND s.z BETWEEN 0.3 AND 0.4' 
res = SDSS.query_sql(query)
print(res)
```

```python
data = res
x = data['redshift']
y = data['petroMag_g']
plt.plot(x, y, color = 'orchid')
plt.ylabel("petroMag_g")
plt.xlabel("z")
plt.title("petroMag_g and z")
plt.show()
```
